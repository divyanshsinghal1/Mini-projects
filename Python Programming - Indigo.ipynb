```python
from sklearn.datasets import load_boston
bs = load_boston()
print(bs.DESCR)
```

```python
import pandas as pd
boston = pd.DataFrame(bs.data, columns = bs.feature_names)
boston['MEDV'] = bs.target
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(boston.corr(),annot=True,cmap='Blues')
plt.show()
```

```python
from sklearn.model_selection import train_test_split as split
train_bhp, test_bhp = split(boston, test_size = .30, random_state = 12)
```

```python
print(train_bhp.shape[0],test_bhp.shape[0])
```

```python
' + '.join(train_bhp)
```

```python
formula = 'MEDV ~ ' +' + '.join(train_bhp.drop(columns='MEDV'))
```

```python
from statsmodels.formula.api import ols
mod = ols(formula,data=train_bhp)
lm_fit=mod.fit()
lm_fit.summary()
```

```python
y_cap = lm_fit.predict(train_bhp.drop(columns='MEDV'))
y = train_bhp.MEDV
residual = y_cap - y
```

```python
plt.figure(figsize=(10,5))
sns.residplot(y_cap, 'MEDV', data=train_bhp, lowess=True, line_kws={'color':'red'})
plt.ylabel('Residual')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
drop_cols = ['MEDV', 'INDUS', 'AGE', 'TAX', 'RAD', 'NOX', 'DIS']
# VIF
x_data = train_bhp.drop(columns=drop_cols)
x_data['intercept'] = 1
VIF = pd.DataFrame()
VIF['variables'] = x_data.columns
VIF['vif']=[vif(x_data.values,i) for i in range(x_data.shape[1])]
VIF = VIF.set_index('variables').drop(index='intercept').T
# Model
form = 'MEDV ~ '+ '+ '.join(train_bhp.drop(columns=drop_cols))
mod_lm = ols(form, data = train_bhp).fit()
print(mod_lm.rsquared,mod_lm.rsquared_adj)
```

```python
inf = mod_lm.get_influence()
```

```python
c,p=inf.cooks_distance
```

```python
plt.stem(range(train_bhp.shape[0]),c)
plt.show()
```

```python
n = train_bhp.shape[0]
p = train_bhp.drop(columns='MEDV').shape[1]
tresh = 4/(n-p-1)
```

```python
final_train = train_bhp[~(c > tresh)].drop(columns=['INDUS', 'AGE', 'TAX', 'RAD', 'NOX', 'DIS'])
```
