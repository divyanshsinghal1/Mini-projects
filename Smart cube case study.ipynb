```python
data = data.drop(['Customer ID'],axis=1)
```

```python
data['Inflight wifi service']=data['Inflight wifi service'].replace([0],[None])
data['Departure/Arrival time convenient']=data['Departure/Arrival time convenient'].replace([0],[None])
data['Cleanliness']=data['Cleanliness'].replace([0],[None])
data['Ease of Online booking']=data['Ease of Online booking'].replace([0],[None])
data['Gate location']=data['Gate location'].replace([0],[None])
data['Food and drink']=data['Food and drink'].replace([0],[None])
data['Online boarding']=data['Online boarding'].replace([0],[None])
data['Seat comfort']=data['Seat comfort'].replace([0],[None])
data['Inflight entertainment']=data['Inflight entertainment'].replace([0],[None])
data['On-board service']=data['On-board service'].replace([0],[None])
data['Leg room service']=data['Leg room service'].replace([0],[None])
data['Baggage handling']=data['Baggage handling'].replace([0],[None])
data['Checkin service']=data['Checkin service'].replace([0],[None])
data['Inflight service']=data['Inflight service'].replace([0],[None])
```

```python
data = data.dropna()
```

```python
data['Inflight wifi service']=data['Inflight wifi service'].astype(float)
data['Departure/Arrival time convenient']=data['Departure/Arrival time convenient'].astype(float)
data['Cleanliness']=data['Cleanliness'].astype(float)
data['Ease of Online booking']=data['Ease of Online booking'].astype(float)
data['Gate location']=data['Gate location'].astype(float)
data['Food and drink']=data['Food and drink'].astype(float)
data['Online boarding']=data['Online boarding'].astype(float)
data['Seat comfort']=data['Seat comfort'].astype(float)
data['Inflight entertainment']=data['Inflight entertainment'].astype(float)
data['On-board service']=data['On-board service'].astype(float)
data['Leg room service']=data['Leg room service'].astype(float)
data['Baggage handling']=data['Baggage handling'].astype(float)
data['Checkin service']=data['Checkin service'].astype(float)
data['Inflight service']=data['Inflight service'].astype(float)
```

```python
data.info()
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data.drop('satisfaction',axis=1),data['satisfaction'],test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
predictions = logmodel.predict(x_test)
```

```python
score = logmodel.score(x_test, y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = x_train, y = y_train, cv = 10)
accuracies.mean()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'n_estimators': [10,50,100], 'criterion': ['entropy']},
              {'n_estimators': [10,50,100], 'criterion': ['gini']}]
grid_search = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)
grid_search = grid_search.fit(x_train, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
```
